```python
import requests
from bs4 import BeautifulSoup
import string
```

```python
import numpy as np
```

```python
Enter_input = string.capwords(input('Search: '))
lists = Enter_input.split()
word = "_".join(lists)

url = 'https://en.wikipedia.org/wiki/'+word

def wikiscrip(urls):
    
    #Getting and parse html content from a Wikipedia page
    
    url_req = requests.get(urls)
    soup = BeautifulSoup(url_req.content, 'html.parser')
    
    #Extract article title
    
    print(soup.title.string)
                 
    # Extract article text for each paragraph with their respective headings
    # and Map those headings to their respective paragraphs in the dictionary
    
    details = soup('table', {'class': 'infobox'})
    
    for i in details:
        
        h = i.find_all('tr')
        for j in h:
            heading = j.find_all('th')
            data = j.find_all('td')
            if heading is not None and data is not None:
                for x,y in zip(heading, data):
                    print('{} :: {}'.format(x.text, y.text))
                    print('......................')
        
        #Extract article text for each paragraph with their respective
        
        for i in range(1,4):
            print(soup('p')[i].text)
            
        
        print('......................')
     
    
    #Printing every link that redirects to another Wikipedia page
    
    for link in soup.find_all('a'):
        print(link.get('href'))
            
    print(soup.title.string)    


wikiscrip(url)
```
